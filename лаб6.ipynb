В файле найти наименьшее число.

```python
def find_min_in_file(filename):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            # Читаем всё содержимое, делим на отдельные элементы по пробелам/переносам строк
            content = file.read().split()
            
            # Преобразуем строковые элементы в числа (int или float)
            numbers = [float(x) for x in content]
            
            if not numbers:
                return "Файл пуст"
            
            return min(numbers)
    except FileNotFoundError:
        return f"Файл {filename} не найден"

# Пример использования:
# Создадим тестовый файл
with open("numbers.txt", "w") as f:
    f.write("25.5 12 -3.4 8 0 44")

min_val = find_min_in_file("numbers.txt")
print(f"Наименьшее число в файле: {min_val}")
```

Дана последовательность символов в файле. Преобразовать файл, заменив все вопросительные знаки нулями

```python
def replace_question_marks(filename):
    try:
        # 1. Читаем данные из файла
        with open(filename, 'r', encoding='utf-8') as file:
            content = file.read()
        
        # 2. Заменяем '?' на '0'
        modified_content = content.replace('?', '0')
        
        # 3. Перезаписываем файл обновленным текстом
        with open(filename, 'w', encoding='utf-8') as file:
            file.write(modified_content)
            
        print("Файл успешно преобразован.")
    except FileNotFoundError:
        print(f"Файл {filename} не найден")

# Пример использования:
with open("text.txt", "w", encoding='utf-8') as f:
    f.write("Привет! Как дела? Сколько время? 2+2=?")

replace_question_marks("text.txt")

# Проверим результат:
with open("text.txt", "r", encoding='utf-8') as f:
    print("Результат:", f.read())
```

Дан файл, компоненты которого являются целыми числами. Записать во второй файл числа, являющиеся точными квадратами четных чисел

```python
import math

def filter_even_squares(input_file, output_file):
    try:
        with open(input_file, 'r', encoding='utf-8') as f_in:
            # Считываем все числа из первого файла
            numbers = [int(x) for x in f_in.read().split()]
        
        valid_numbers = []
        for num in numbers:
            if num >= 0: # Квадрат целого числа не может быть отрицательным
                root = math.isqrt(num) # Находит целую часть квадратного корня
                # Проверяем, является ли число точным квадратом и чётный ли корень
                if root * root == num and root % 2 == 0:
                    valid_numbers.append(str(num))
        
        # Записываем результат во второй файл (через пробел)
        with open(output_file, 'w', encoding='utf-8') as f_out:
            f_out.write(" ".join(valid_numbers))
            
        print(f"Отобранные числа записаны в файл {output_file}")
    except FileNotFoundError:
        print(f"Файл {input_file} не найден")

# Пример использования:
# Создаем исходный файл с набором целых чисел
with open("source_ints.txt", "w") as f:
    f.write("4 9 16 25 36 50 64 -4 0") 
    # 4 (2²-ок), 9 (3²-нет), 16 (4²-ок), 25 (5²-нет), 36 (6²-ок), 64 (8²-ок), 0 (0²-ок)

filter_even_squares("source_ints.txt", "result_squares.txt")

# Проверим результат:
with open("result_squares.txt", "r") as f:
    print("Числа во втором файле:", f.read())
```
