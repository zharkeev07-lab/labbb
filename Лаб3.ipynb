1) Даны внутренний и внешний радиусы кольца. Найти площадь
кольца.


```python
import math

def ring_area(r_inner, r_outer):
    if r_inner >= r_outer:
        return "Ошибка: внутренний радиус должен быть меньше внешнего!"
    
    area = math.pi * (r_outer**2 - r_inner**2)
    return area

# Пример использования:
r_in = 3.0
r_out = 5.0
print(f"Площадь кольца: {ring_area(r_in, r_out):.2f}")
```

2) Треугольник задан координатами вершин. Найти периметр и площадь треугольника.

```python
import math

def triangle_properties(x1, y1, x2, y2, x3, y3):
    # Находим длины сторон
    a = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    b = math.sqrt((x3 - x2)**2 + (y3 - y2)**2)
    c = math.sqrt((x1 - x3)**2 + (y1 - y3)**2)
    
    perimeter = a + b + c
    
    # Площадь по координатам (формула Гаусса)
    area = 0.5 * abs(x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2))
    
    return perimeter, area

# Пример использования (вершины (0,0), (4,0), (0,3))
p, s = triangle_properties(0, 0, 4, 0, 0, 3)
print(f"Периметр: {p:.2f}, Площадь: {s:.2f}")
```

3) Даны длины сторон параллелепипеда. Найти объем и все его диагонали, сравнить их и вывести длину минимальной диагонали.

```python
import math

def parallelepiped_properties(a, b, c):
    volume = a * b * c
    
    # Диагонали граней
    d_ab = math.sqrt(a**2 + b**2)
    d_bc = math.sqrt(b**2 + c**2)
    d_ac = math.sqrt(a**2 + c**2)
    # Пространственная диагональ
    d_space = math.sqrt(a**2 + b**2 + c**2)
    
    all_diagonals = [d_ab, d_bc, d_ac, d_space]
    min_diagonal = min(all_diagonals)
    
    return volume, all_diagonals, min_diagonal

# Пример использования:
v, diags, min_d = parallelepiped_properties(3, 4, 5)
print(f"Объем: {v}")
print(f"Все уникальные диагонали: {[round(d, 2) for d in diags]}")
print(f"Минимальная диагональ: {min_d:.2f}")
```

Найти площадь треугольника по известным длинам трех сторон. Вывести максимальную высоту

```python
import math

def triangle_by_sides(a, b, c):
    # Проверка существования треугольника
    if (a + b <= c) or (a + c <= b) or (b + c <= a):
        return "Треугольник с такими сторонами не существует"
    
    # Формула Герона
    p = (a + b + c) / 2
    area = math.sqrt(p * (p - a) * (p - b) * (p - c))
    
    # Максимальная высота опускается на минимальную сторону
    min_side = min(a, b, c)
    max_height = (2 * area) / min_side
    
    return area, max_height

# Пример использования:
s, h_max = triangle_by_sides(13, 14, 15)
print(f"Площадь: {s:.2f}, Максимальная высота: {h_max:.2f}")
```

 Сравнить между собой 4 числа

```python
def compare_four_numbers(n1, n2, n3, n4):
    numbers = [n1, n2, n3, n4]
    
    maximum = max(numbers)
    minimum = min(numbers)
    sorted_numbers = sorted(numbers)
    
    return maximum, minimum, sorted_numbers

# Пример использования:
num_max, num_min, num_sorted = compare_four_numbers(15, -3, 42, 0)
print(f"Максимальное: {num_max}")
print(f"Минимальное: {num_min}")
print(f"Числа по порядку: {num_sorted}")
```
