Построить график функции $y = \frac{x}{2x^2 + 3x - 1}$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Генерируем массив точек x
x = np.linspace(-5, 5, 1000)

# Считаем знаменатель и саму функцию
denominator = 2 * x**2 + 3 * x - 1
y = x / denominator

# Маскируем экстремальные значения около асимптот для красивого отображения
y[abs(y) > 10] = np.nan

# Строим график
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='royalblue', linewidth=2, label='y = x / (2x^2 + 3x - 1)')

# Оформление осей и сетки
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.title('График функции', fontsize=14)
plt.xlabel('X')
plt.ylabel('Y')
plt.ylim(-5, 5)  # Ограничиваем по Y для наглядности
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()

plt.show()
```

Нарисовать шахматную доску 4∙4.

```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем сетку 4x4 с чередованием (0 - черный, 1 - белый)
# np.indices создает матрицы индексов строк и столбцов, их сумма дает шахматный паттерн
board = (np.indices((4, 4)).sum(axis=0) % 2)

plt.figure(figsize=(5, 5))
# Отображаем матрицу в черно-белой цветовой гамме
plt.imshow(board, cmap='gray', extent=[0, 4, 0, 4])

# Настраиваем сетку и деления, чтобы было похоже на доску
plt.xticks(range(5))
plt.yticks(range(5))
plt.title('Шахматная доска 4x4', fontsize=14)

plt.show()
```

Нарисовать бабочку.

```python
import numpy as np
import matplotlib.pyplot as plt

# Задаем диапазон параметра t для плавных линий
t = np.linspace(0, 12 * np.pi, 2500)

# Параметрическое уравнение кривой бабочки
r = np.exp(np.sin(t)) - 2 * np.cos(4 * t) + np.sin((2 * t - np.pi) / 24) ** 5
x = r * np.sin(t)
y = r * np.cos(t)

plt.figure(figsize=(6, 6))
# Строим контур фиолетовым цветом
plt.plot(x, y, color='darkorchid', linewidth=1.5)

# Скрываем оси координат для чистоты рисунка
plt.axis('off')
plt.title('Математическая бабочка', fontsize=14)

plt.show()
```
