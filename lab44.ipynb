```python
# Инициализация массива для хранения 20 чисел
x = [0] * 21  # Индексация с 1 до 20, поэтому размер 21
x[1] = x[2] = x[3] = 1

# Заполняем массив по формуле x_i = x_{i-1} + x_{i-3}
for i in range(4, 21):
    x[i] = x[i-1] + x[i-3]

# Вычисляем сумму
total_sum = 0
for i in range(1, 21):
    total_sum += x[i] / (2**i)

print(f"Результат суммы: {total_sum}")
```

Нахождение суммы $\sum_{i=1}^{20} \frac{x_i}{2^i}$

В этой задаче важно сначала получить значения всех x_i, а потом вычислить сумму.  
x = [0] * 21: Мы создаем список из 21 нуля. Нам нужно 21 место (индексы от 0 до 20), чтобы удобно обращаться к x[1]...x[20], игнорируя нулевой индекс.  
x[1] = x[2] = x[3] = 1: Базовые значения последовательности согласно условию.  
for i in range(4, 21): x[i] = x[i-1] + x[i-3]: Это основной алгоритм (рекуррентная формула).  
 Мы заполняем массив, где каждое новое число — это сумма двух предыдущих (отстоящих на 1 и на 3 позиции назад).  
 total_sum += x[i] / (2^i): Мы проходим циклом по всем элементам и прибавляем к сумме дробь, где 2^i — это возведение двойки в степень i.

Симметричный разворот массива  

import random: Импортируем модуль для генерации случайных чисел.  
[random.randint(1, 100) for _ in range(n)]: Генератор списков. Он создает массив из n случайных целых чисел от 1 до 100.  
arr.reverse(): Это встроенный метод Python, который «переворачивает» список на месте (первый становится последним, второй — предпоследним и так далее), что полностью отвечает требованию симметричной перестановки.

```python
n = 30
arr = [random.randint(1, 100) for _ in range(n)]

print("Числа, индексы которых содержат цифру 2:")
for i in range(n):
    if '2' in str(i):
        print(f"Индекс {i}: значение {arr[i]}")
```

Вывод чисел, в индексе которых есть цифра "2"

for i in range(n):: Мы проходим по всем индексам массива.  
if '2' in str(i):: Это самый изящный способ проверки. Мы превращаем число-индекс в строку (str(i)), а затем проверяем наличие символа '2' в этой строке. Если индекс, например, 12 или 25, условие вернет True.

```python
import random

n = 5
k = 2  # Например, строка 2 (индексация с 0)
matrix = [[random.randint(1, 50) for _ in range(n)] for _ in range(n)]

# 1. Обнуление строк выше k
for i in range(k):
    for j in range(n):
        matrix[i][j] = 0

# 2. Сбор элементов ниже строки k, сортировка и возврат
below_elements = []
for i in range(k + 1, n):
    below_elements.extend(matrix[i])

below_elements.sort(reverse=True)

# Заполняем обратно
idx = 0
for i in range(k + 1, n):
    for j in range(n):
        matrix[i][j] = below_elements[idx]
        idx += 1
```

Работа с матрицей $n \times n$

Вложенные списки ([[...]]): Матрица в Python — это список, внутри которого лежат другие списки (строки).  
Обнуление: Мы используем двойной цикл (по строкам i и столбцам j) и задаем matrix[i][j] = 0 для тех строк, где i < k.  
Сортировка: Чтобы отсортировать часть матрицы, мы сначала "вытягиваем" нужные строки в один плоский список below_elements. Используем метод .sort(reverse=True) для сортировки по убыванию, а затем снова вставляем эти числа в матрицу с помощью двойного цикла.

```python
import math

matrix = [[random.randint(-10, 10) for _ in range(4)] for _ in range(4)]

# 1. Главная диагональ (элементы [0][0], [1][1], [2][2], [3][3])
prod = 1
for i in range(4):
    prod *= matrix[i][i]
geom_mean = prod**(1/4)

# 2. Поиск макс. элемента побочной диагонали ([0][3], [1][2], [2][1], [3][0])
side_diag = [matrix[i][3-i] for i in range(4)]
max_val = max(side_diag)

# Замена
for i in range(4):
    if matrix[i][3-i] == max_val:
        matrix[i][3-i] = geom_mean
        break

# 3. Обнуление положительных элементов
for i in range(4):
    for j in range(4):
        if matrix[i][j] > 0:
            matrix[i][j] = 0
```

Среднее геометрическое и замена

Среднее геометрическое: Для набора из 4 чисел это корень 4-й степени из их произведения: $\sqrt[4]{x_1 \cdot x_2 \cdot x_3 \cdot x_4}$. В коде это записано как prod(1/4).  
Побочная диагональ: В матрице 4x4 элементы побочной диагонали имеют индексы [0][3], [1][2], [2][1], [3][0]. Общая формула индекса столбца для строки i — это 3-i.  
if matrix[i][j] > 0: matrix[i][j] = 0: Мы проходим по всей матрице двойным циклом for, проверяем каждый элемент. Если он положительный, превращаем его в 0.
